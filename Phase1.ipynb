```python
import numpy as np
import pandas as pd 
data=pd.read_csv('Phase1.csv')
data.info() # Voir le type de données du dataframe 
```

```python
data.describe() # Voir les statistiques 
# On remarque que les colonnes 'urgent' 'num_outbound_cmds' 'is_host_login' 
# contiennent uniquement des 0, avec un min = 0 et max = 0 et mean = 0
```

```python
# Vérifier si on a des valeurs manquantes
data.isnull().sum()
```

```python
# Vérifier si on a des données en redondance 
data.duplicated()
```

```python
# Suppression de la colonne 'urgent' car elle contient que des '0'
data=data.drop('urgent',axis='columns')
data=data.drop('num_outbound_cmds',axis='columns')
data=data.drop('is_host_login',axis='columns')
data.shape # Vérifier si elles ont été supprimées 
```

```python
# Conversion des valeures qualitatives en quantitatives
data=pd.get_dummies(data)
data.head() #Vérifier si la conversion a été faite
```

```python
X = data.iloc[:,:-2] # Tout récuperer sauf les 2 derniéres colonnes 
Y = data.iloc[:,-2:] # Récuperer les 2 derniéres colonnes
from sklearn.preprocessing import scale
X=scale(X) 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,stratify=Y)
```

```python
def cal(y_test,y_pred): # La fonction prend en paramétre 2 numpy arrays
    vp,vn,fp,fn=0,0,0,0
    for i in range(0,len(y_test)): # Boucle qui parcourt les 2 numpy arrays vu qu'ils ont la même taille 
        #Vrai_Positif
        if((y_test[i,0]==1)&(y_pred[i,0]==1)):
            vp+=1
        #Vrai_Négatif
        elif((y_test[i,1]==1)&(y_pred[i,1]==1)):
            vn+=1
        #Faux_Positif
        elif((y_test[i,1]==1)&(y_pred[i,0]==1)):
            fp+=1
        #Faux_Négatif
        elif((y_test[i,0]==1)&(y_pred[i,1]==1)):
            fn+=1
        
    return vp,vn,fp,fn # Retourner les résultats trouvés
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
params = {'n_neighbors':[3,5,7,9,11,13,15,17],'metric':['euclidean','manhattan']}
grid=GridSearchCV(KNeighborsClassifier(),params,cv=5) # Chercher les meilleurs paramétres
g=grid.fit(X_train,Y_train)
g.best_params_ # Afficher les meilleurs paramétres
```

```python
Knn= KNeighborsClassifier(n_neighbors=3,metric='manhattan')
Knn.fit(X_train,Y_train) 
y_pred=Knn.predict(X_test)
```

```python
type(y_pred)
#On remarque que le résultat de y_pred est de type numpy array, mais que Y_test est de type dataframe,
#donc pour utiliser notre  fonction on doit convertir pour comparer entre les mêmes data types
```

```python
#Conversion du dataframe vers numpy array pour pouvoir comparer entre y_test et y_pred dans la fonction
y_test=Y_test.to_numpy()
```

```python
vp,vn,fp,fn=cal(y_test,y_pred)
prec = vp/(vp+fp)
rap = vp/(vp+fn)
f = (2*prec*rap)/(rap+prec)
print("Precision = "+str(prec)+"\nRappel = "+str(rap)+"\nF1_score = "+str(f))
```
